```python

# Create the DataFrame with columns : Date,PRoduct_A,PRoduct_B,PRoduct_C,PRoduct_D
year = 2024
product_sales_Df = pd.DataFrame({
    'Date': pd.date_range(start=f"{year}-01-01", end=f"{year}-12-30"),
    'Product_A':product_A,
    'Product_B':product_B,
    'Product_C':product_C,
    'Product_D':product_D,
})
```
